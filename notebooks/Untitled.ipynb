
## 3) Q-learning ( Elian )


For experiments with tabular Q-learning the action-vale function was initialized to zero. The actions are chosen from the Q-table or with e-greedy scheduling through a exploration probability (epsilon) that is updated through a exponential decay formula en each episode, this probability is compared with a random float. If the random float is less than the exploration probability the agent selects a random action else he exploits his knowledge using the bellman equation .

The Q-learning function updates the Q-table, this procedure depends of learning rate and gamma values. For this experiments section were used a GridSearch method in order to seek the combination of parameters that make the agent achieve the goal the best of possible.


```python
from qlearning import *
```

### 3.1) Tabular & Deterministic

```python
report_sarsa(stochastic = False)
```

```python
env = gym.make('FrozenLake8x8-v1', is_slippery=False)
Q_table, stats = qlearning_tabular(env, episodes=1000, epsilon_decay=0.01, gamma_=1.0, alpha=0.001 )

plt.plot(stats)
plt.show()
```

```python
print_policy_q(Q_table)
```

### 3.2) Tabular & Stochastic


```python
report_sarsa(stochastic = True)
```

```python
env = gym.make('FrozenLake8x8-v1', is_slippery=True)
Q_stable, stats = qlearning_tabular(env, episodes=5000, epsilon_decay=0.001, gamma_=0.99, alpha=0.1 )

plt.plot(stats)
plt.show()
```

```python
print_policy_q(Q_stable)
```
